## Make some plots!

```python
impath = savedir
hsc_filters = ["g", "r", "i", "z"]
mer_filters = ["N540", "N708"]
filters = hsc_filters+ mer_filters

srcs = np.genfromtxt(cfile)
```

For one source:

```python
ra, dec = srcs[1]
sc = coordinates.SkyCoord (ra, dec, unit='deg' )
cname = produce_merianobjectname(sc)
```

Load images

```python
cutouts = dict([(f, fits.open(os.path.join(impath, "hsc", f"{cname}_HSC-{f}.fits"))) for f in hsc_filters])
cutouts.update(dict([(f, fits.open(os.path.join(impath, "merian", f"{cname}_{f}_merim.fits"))) for f in mer_filters]))
psfs = dict([(f, fits.open(os.path.join(impath, "hsc", f"{cname}_HSC-{f}_psf.fits"))[0].data) for f in hsc_filters])
psfs.update(dict([(f, fits.open(os.path.join(impath, "merian", f"{cname}_{f}_merpsf.fits"))[0].data) for f in mer_filters]))

cutout_headers = dict([(f, cutouts[f][1].header) for f in filters])
cutouts = dict([(f, cutouts[f][1].data) for f in filters])

psf_shape = np.max(np.array([i.shape for i in psfs.values()]), 0)
cutout_shape = np.min(np.array([i.shape for i in cutouts.values()]), 0)
```

Make an image

```python
image = make_lupton_rgb(cutouts["i"], cutouts["r"], cutouts["g"], Q=8, stretch=0.5)
fig, ax = plt.subplots(1,3, figsize = (20,5))

ax[0].set_title("gri")
ax[0].imshow(image)

ax[1].set_title("N540")
ax[1].imshow(cutouts["N540"], cmap="plasma", norm = colors.AsinhNorm())

ax[2].set_title("N708")
ax[2].imshow(cutouts["N708"], cmap="plasma", norm = colors.AsinhNorm())

plt.show()
```
